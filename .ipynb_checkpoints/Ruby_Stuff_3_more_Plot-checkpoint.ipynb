# COLORS 
## https://www2.uni-hamburg.de/Wiss/FB/15/Sustainability/schneider/gnuplot/colors.htm

```ruby
require 'gnuplotrb'
include GnuplotRB

cartesian = Plot.new(
  ['x*sin(x)', with: 'lines', lt: { rgb: 'coral', lw: 3 }],
  xrange: -10..10,
  title: 'Math function example', 
  ylabel: 'x',
  xlabel: 'x*sin(x)'
)
```

```ruby
Plot.new(
  ['abs(sin(3*t))', with: 'filledcurves', lt: { rgb: 'red' }],
  title: 'Plot in polar example',
  polar: true,
  samples: 500
)
```

```ruby
Plot.new(
  ['abs(sin(4*t))', with: 'filledcurves', lt: { rgb: 'red' }],
  ['abs(sin(3*t))', with: 'filledcurves', lt: { rgb: 'yellow' }],
  ['abs(sin(2*t))', with: 'filledcurves', lt: { rgb: 'blue' }],
  title: 'Plot in polar example',
  polar: true,
  samples: 500
)
```

```ruby
Plot.new(
  ['1.5*cos(t) - cos(30*t), 1.5*sin(t) - sin(30*t)', title: 'Parametric curve', lw: 2],
  title: 'Parametric plot example',
  parametric: true,
  samples: 1000,
  xrange: -3..3,
  yrange: -3..3
)
```
